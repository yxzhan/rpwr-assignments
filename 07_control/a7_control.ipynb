## Task 1 - TF Referee

Write a python node that publishes waypoints as TF frames, and recognizes when the turtlebot arrived there. The node should have the following functionality:

1. Publish a list of at least 2 transforms as goal waypoints for the turtlebot.

2. When the TF frame of the Turtlebot gets within 20cm range of one of the goal waypoints:
  * The referee node publishes a message of type `std_msgs/String` to the `/goal_collected` topic.
  * The referee node **stops** publishing that goal frame to.
  
Start off by publishing just one frame and check for the condition of point 2, so you have a solid foundation for task 2.

## Task 2 - Navigate through a door

Your turtle is in the hallway and wants to enter a room. Choose a single waypoint 1.00 meter in the room and use that direction in your calculation for `/cmd_vel`. Make the robot autonomously navigate through that open door into the room to the goal waypoint. The use of the `ActionLib` library or similar libraries are not allowed.

Hint: Use the `lookup_transform` as heuristic to calculate the direction of the `cmd_vel`.
